```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags, kron, lil_matrix
from scipy.sparse.linalg import eigsh

class KaluzaKleinSimulator:
    def __init__(self, R=0.1, N_H3=5, N_S1=64, include_potential=True, num_modes=20):
        """
        Initialisation du simulateur avec des paramètres réduits pour tests.
        """
        self.R = R
        self.N_H3 = N_H3
        self.N_S1 = N_S1
        self.include_potential = include_potential
        self.num_modes = num_modes

        self.dim_total = (self.N_H3**3) * self.N_S1
        print(f"Dimension totale : {self.dim_total}")

        self.X = np.linspace(0, 2 * np.pi, self.N_H3)
        self.Theta = np.linspace(0, 2 * np.pi, self.N_S1)

    def hamiltonian_H3(self):
        """Hamiltonien 3D sur une grille régulière."""
        size = self.N_H3
        diag = -6 * np.ones(size**3)
        off_diag = np.ones(size**3 - 1)

        # Matrices 1D pour chaque direction (x, y, z)
        Hx = diags([off_diag, diag, off_diag], [-1, 0, 1], shape=(size, size))
        Hy = Hx.copy()
        Hz = Hx.copy()

        # Construction du Laplacien 3D via Kronecker
        H_H3 = kron(kron(Hx, np.eye(size)), np.eye(size)) + \
               kron(kron(np.eye(size), Hy), np.eye(size)) + \
               kron(kron(np.eye(size), np.eye(size)), Hz)

        return H_H3

    def laplacian_S1(self):
        """Laplacien sur S¹ avec conditions périodiques optimisées."""
        diagonals = [-2 * np.ones(self.N_S1), np.ones(self.N_S1 - 1), np.ones(self.N_S1 - 1)]
        offsets = [0, -1, 1]
        L = diags(diagonals, offsets, shape=(self.N_S1, self.N_S1)).tolil()
        L[0, -1] = L[-1, 0] = 1
        return L.tocsr() / (self.R**2 * (2 * np.pi / self.N_S1)**2)

    def build_hamiltonian(self):
        """Construction de l'Hamiltonien total."""
        H_H3 = self.hamiltonian_H3()
        H_S1 = self.laplacian_S1()

        H = kron(H_H3, np.eye(self.N_S1)) + kron(np.eye(self.N_H3**3), H_S1)

        if self.include_potential:
            print("Ajout du potentiel...")
            V_h3 = np.cos(self.X).reshape(-1, 1)
            V_s1 = np.cos(self.Theta)
            V_flat = np.kron(V_h3.flatten(), V_s1).flatten()
            V_flat = np.tile(V_flat, self.N_H3**2)
            V = diags(V_flat, 0, shape=(self.dim_total, self.dim_total))
            H += V

        return H

    def analyze_S1_spectrum(self):
        """Analyse du spectre de S¹ seul."""
        H_S1 = self.laplacian_S1()
        eigvals_S1 = eigsh(H_S1, k=min(self.N_S1-2, self.num_modes), which='SA', return_eigenvectors=False)
        return np.sort(eigvals_S1), np.diff(np.sort(eigvals_S1))

    def detect_kk_towers(self):
        """Détection des tours de Kaluza-Klein avec visualisation."""
        print("Construction de l'Hamiltonien complet...")
        H = self.build_hamiltonian()

        if self.num_modes >= self.dim_total:
            print("⚠️ Attention : Le nombre de modes demandés est trop grand !")
            return

        print(f"Calcul des {self.num_modes} valeurs propres pour H...")
        eigvals = eigsh(H, k=self.num_modes, which='SA', return_eigenvectors=False)
        eigvals = np.sort(eigvals)
        delta_E = np.diff(eigvals)

        eigvals_S1, delta_E_S1 = self.analyze_S1_spectrum()

        plt.figure(figsize=(15, 10))
        plt.subplot(221)
        plt.plot(eigvals, 'o-', label=f'R={self.R}')
        plt.xlabel('Mode')
        plt.ylabel('Énergie')
        plt.title('Énergies propres (H³ × S¹)')
        plt.legend()

        plt.subplot(222)
        plt.stem(delta_E, linefmt='C1-', markerfmt='C1o', label='ΔE observé')
        plt.yscale('log')
        plt.xlabel('Indice')
        plt.ylabel('ΔE')
        plt.title('Espacement entre les modes (H³ × S¹)')
        plt.legend()

        plt.subplot(223)
        plt.plot(eigvals_S1, 'o-', label=f'S¹ seul, R={self.R}')
        plt.xlabel('Mode')
        plt.ylabel('Énergie')
        plt.title('Énergies propres (S¹ seul)')
        plt.legend()

        plt.subplot(224)
        plt.stem(delta_E_S1, linefmt='C2-', markerfmt='C2o', label='ΔE observé (S¹)')
        plt.yscale('log')
        plt.xlabel('Indice')
        plt.ylabel('ΔE')
        plt.title('Espacement entre les modes (S¹ seul)')
        plt.legend()

        plt.tight_layout()
        plt.show()
        print("Simulation terminée.")

if __name__ == "__main__":
    print("Simulation avec potentiel :")
    simulator = KaluzaKleinSimulator(R=0.1, N_H3=5, N_S1=64, include_potential=True, num_modes=20)
    simulator.detect_kk_towers()
```
